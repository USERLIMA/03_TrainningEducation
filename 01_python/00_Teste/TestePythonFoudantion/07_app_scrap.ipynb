```python
# Import das Libs
import pandas as pd
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError
from bs4 import BeautifulSoup
```

```python
urlBase = 'https://www.imdb.com/chart/top/'

# Tratando os erros

try:
    html = urlopen(urlBase)
except HTTPError as e:
    # Erros de HTTP
    print(e)
except URLError as e:
    # URL Errada
    print('A URL informada, está errada!')
```

```python
bs = BeautifulSoup(html, 'lxml')
```

```python
movies = bs.select('.lister-list tr') # selecionando todos os elementos 'tr' dentro do elemento 'lister-list'
```

```python
# Armazenando as informações dentro de listas
titles = []
directors_writers = []
years = []
ratings = []
```

```python
# Recuperando as informações desejadas e armazenando nas listas que criamos acima
for movie in movies:
    titles.append(movie.find('td', class_='titleColumn').find('a').get_text())
    directors_writers.append(movie.find('td', class_='titleColumn').find('a')['title'])
    years.append(movie.find('td', class_='titleColumn').find('span').get_text()[1:5])
    ratings.append(movie.find('td', class_='imdbRating').find('strong').get_text())
```

Criando um dataframe utilizando a dependencia pandas
Abaixo, o dataframe é criado a partir do resultado da extração de dados obtida com o scraping. Observe o código abaixo:

```python
# Criando um conjunto de informações
df = pd.DataFrame({
    "titulo": titles,
    "ano de lançamento": years,
    "classificação": ratings,
    "diretores e roteiristas": directors_writers
})
```

```python
# Salvando o Dataframe para CSV
df.to_csv('WS_Filmes.csv')
```
