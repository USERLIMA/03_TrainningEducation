# Introdução ao Matplotlib

É uma lib Python para trabalhar com gráficos 2D.

# Tipos de Plotagem

- Gráfico de linha
- Contorno e Pseudocolor
- Histogramas

```python
# importando a lib

import matplotlib.pyplot as plt
```

# Linha Vertical

## axvline

- Sintaxe básica - plt.axvline(x=0, ymin=0, ymax=1, kwargs)
**kwargs são as minhas propriedades da linha

```python
# Configurações da minha linha vertical

plt.axvline(0.2, 0, 1, label = 'Minha Linha Azul')
plt.legend() # Mostrar o Rótulo/Etiqueta/Label
plt.show() # Mostrar o Gráfico
```

# Linha Horizontal
## Sintaxe - plt.axhline(y = 0, xmin = 0, xmax=1, ** kwargs)

```python
ypoints = 0.1
plt.axhline(ypoints, 0, 1, label= 'Minha Linha Horizontal')
plt.legend()
plt.show()
```

# Plotando múltiplas linhas

```python
xpoints = [0.2, 0.4, 0.6]

for p in xpoints:
    plt.axvline(p, label=' Vertical Lines')
    
plt.legend()
plt.show()
```

```python
# Mudando as configurações do meu gráfico

xpoints = [0.2, 0.4, 0.6]
colors = ['g', 'c', 'm']

# p = position em xpoints
# c = position em colors
for p, c in zip(xpoints, colors):
    plt.axvline(p, label='line: {}'.format(p), color=c)
# Usamos a variavel 'p' em for para criar um loop que pegue todos os pontos para o vetor, para que nao tenhamos que colar todos os pontos
# {} serve para representar as informações que vou mostrar 

plt.legend()
plt.savefig('vertical_lines.png')
plt.show()
```

```python
# Gráfico de forma horizontal
ypoints = [0.2, 0.4, 0.6, 0.68, 2]
colors = ['b', 'k', 'y', 'm',  'k']

for p, c in zip(ypoints, colors):
    plt.axhline(p, label='Line: {}'.format(p), color=c) # ou c=c

plt.legend()
plt.show()
```

#### LEMBRAR DE TENTAR ALTERAR A DESCRIÇÃO DO GRÁFICO

```python
# Gráfico de forma horizontal
ypoints = [0.2, 0.4, 0.6, 0.68, 2]
colors = ['b', 'k', 'y', 'm',  'k']
desc = [
    if ypoints = 0.2:
        print('desc = 0.2', ypoints)
    elif ypoints = 0.4:
        print('desc = 0.4', ypoints)
    elif ypoints = 0.68:
        print('desc = 0.68', ypoints)
    else ypoints = 2:
        print('desc = 2', ypoints)
       ]
for p, c in zip(ypoints, colors):
    plt.axhline(p, label='Line: {}'.format(desc), color=c) # ou c=c

plt.legend()
plt.show()
```

# Subplot Múltiplo

```python
# importando a lib
plt.subplot(2,2,1)
x1 = [40, 50, 60, 70, 80, 90, 100]
y1 = [40, 50, 60, 70, 80, 90, 100]
plt.plot(x1, y1, color="c")

plt.subplot(2,2,2)
x2 = [40, 50, 60, 70, 80, 90, 100]
y2 = [40, 50, 60, 70, 80, 90, 100]
plt.plot(x2, y2, color="m")

plt.subplot(2,2,3)
x3 = [40, 50, 60, 70, 80, 90, 100]
y3 = [40, 50, 60, 70, 80, 90, 100]
plt.plot(x3, y3, color="g")

plt.subplot(2,2,4)
x4 = [40, 50, 60, 70, 80, 90, 100]
y4 = [40, 50, 60, 70, 80, 90, 100]
plt.plot(x4, y4, color="k")
plt.show()
```

# AXIS RANGE
### Limitar os eixos X e Y

```python
# Limitar o eixo X
x1 = [40, 50, 60, 70, 80, 90, 100]
y1 = [40, 50, 60, 70, 80, 90, 100]

plt.plot(x1, y1)

plt.xlim([0, 140]) # Limitando o eixo x até 140
plt.show()
```

```python
# Limitar o eixo Y
x1 = [40, 50, 60, 70, 80, 90, 100]
y1 = [40, 50, 60, 70, 15, 90, 100]

plt.plot(x1, y1)

plt.ylim([0,140]) # Limitando o eixo x até 140
plt.show()
```

```python
# Limitar os eixos X e Y
x1 = [40, 50, 60, 70, 80, 90, 100]
y1 = [40, 50, 60, 70, 80, 90, 100]

plt.plot(x1, y1)

plt.xlim([30,110])
plt.ylim([0,140]) # Limitando o eixo x até 140
plt.show()
```

# Rótulo dos Eixos(Legendas)

```python
x1 = [40, 50, 60, 70, 80, 90, 100]
y1 = [40, 50, 60, 70, 80, 90, 100]

plt.plot(x1, y1)

# Criando as legendas
plt.xlabel('Eixo X') # Legenda do eixo x
plt.ylabel('Eixo Y') # Legenda do eixo y

plt.show()
```

```python
x1 = [40, 50, 60, 70, 80, 90, 100]
y1 = [40, 50, 60, 70, 80, 90, 100]

plt.plot(x1, y1)

# Criando as legendas
plt.xlabel('Eixo X') # Legenda do eixo x
plt.ylabel('Eixo Y') # Legenda do eixo y

plt.clf() # Apaga o gráfico

plt.show()
```
