```python
# import das libs

import pandas as pd
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError 
from bs4 import BeautifulSoup 
```

```python
urlBase = 'https://www.imdb.com/chart/top/'

# Tratando os erros 

try:
    html = urlopen(urlBase)
except HTTPError as e: 
    # Erros de HTTP - as e = erro
    print(e)
except URLError as e:
    print('O servidor/URL não pode ser encontrado ! ')
```

```python
bs = BeautifulSoup(html, 'lxml')
```

```python
movies = bs.select('.lister-list tr') # selecionando todos os elementos 'tr' detro da lista
```

```python
# Armazenando as informações dentro de listas 

titles = []
directors_writers = []
years = []
ratings = []
```

```python
# a = text 
# Recuperando as informações desejadas e armazemando em listas que criamos a cima
for movie in movies:
    titles.append(movie.find('td', class_='titleColumn').find('a').get_text())
    directors_writers.append(movie.find('td', class_='titleColumn').find('a')['title'])
    years.append(movie.find('td', class_='titleColumn').find('span').get_text()[1:5])
    ratings.append(movie.find('td', class_='imdbRating').find('strong').get_text())
```

### criando um data frame - conjunto de informação - utilizando a dependencia pandas abaixo, o data frame é criado a partir do resultado da extração de dados obtida com o scraping

```python
# criando um conjunto de informações 

df = pd.DataFrame({
    'Título':titles,
    'Ano de lançamento': years,
    'Classificação': ratings,
    'Diretores e roteiristas': directors_writers
})
```
