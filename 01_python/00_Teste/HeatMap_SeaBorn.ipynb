# HeatMap com a SeaBorn

## o que é um mapa de calor(HeatMap) - é uma representação visual dos dados que temos no nosso notebook(Jupyter)

```python
# importando as libs

import numpy as np
import seaborn as sb #sns
import matplotlib.pyplot as plt
```

### Importamos o numpy para gerar uma matriz com números aleatórios.

```python
# Gerando a matriz com dados aleatórios

dados = np.random.rand(4,6)
```

```python
# Gerando o Mapa de Calor com os valores de "dados"

heat_map = sb.heatmap(dados)
```

```python
# Removendo as tick labels(Rótulo de Escala) para os eixos x e y
# xticklabels=False remove os labels do eixo x
# yticklabels=False remove os labels do eixo y
heat_map = sb.heatmap(dados, xticklabels=False, yticklabels=False)
```

```python
heat_map = sb.heatmap(dados, xticklabels=False)
```

```python
heat_map = sb.heatmap(dados, yticklabels=False)
```

```python
# Definindo rótulo
heat_map = sb.heatmap(dados)
plt.ylabel("Valores do EIXO Y")
plt.show()
```

```python
# Definindo rótulo
heat_map = sb.heatmap(dados)
plt.xlabel("Valores do EIXO X")
plt.show()
```

```python
# Definindo rótulo
heat_map = sb.heatmap(dados, yticklabels=False, xticklabels=False)
plt.xlabel("Valores do EIXO X")
plt.ylabel("Valores do EIXO Y")
plt.show()
```

```python
# Alterando a cor do HeatMap
heat_map = sb.heatmap(dados, cmap="YlGnBu")
```

```python
# Outros padrões de cores
# https://matplotlib.org/2.0.2/users/colormaps.html
# O nome do padrão é o que está em cima dos gráficos no site acima
heat_map = sb.heatmap(dados, cmap="GnBu")
```

```python
# Adicionando os valores(Annot)
heat_map = sb.heatmap(dados, annot=True, cmap="GnBu")
```

```python
# Trocando o tamanho da fonte
sb.set(font_scale=0.7)
heat_map = sb.heatmap(dados, annot=True, cmap="GnBu")
```

```python
# Ocultando a Color Bar
sb.set(font_scale=0.7)
heat_map = sb.heatmap(dados, annot=True, cmap="GnBu", cbar=False)
```

```python
# Customizar o Gráfico
sb.set(font_scale=0.7)
heat_map = sb.heatmap(
    dados, annot=True, yticklabels=False, xticklabels=False, cmap="GnBu", 
    cbar_kws={'label':'Minha Color Bar'})
plt.xlabel("Valores do EIXO X")
plt.ylabel("Valores do EIXO Y")
plt.show()
```

```python
# Customizar o Gráfico - Trocando a posição da Color Bar
sb.set(font_scale=0.7)
heat_map = sb.heatmap(
    dados, annot=True, yticklabels=False, xticklabels=False, cmap="GnBu", 
cbar_kws={'label':'Minha Color Bar','orientation':'horizontal'})
plt.xlabel("Valores do EIXO X")
plt.ylabel("Valores do EIXO Y")
plt.show()
```
