## Programação orientada à objetos 

Por que partir de uma premissa com esse contexto ? A resposta é que o paradigma de programação orientada à objetos possuiu, por essência, a intenção de aproximar o desenvolvimento de softwares daquilo que conhecemos como mundo real. 

Abaixo podemos observar algumas das vantagens de se programar a partir do paradigma de orientação de objetos: 

1. Confiabilidade: o isolamento entre partes de cada "pedaço" de um sistema que a partir de uma alteração, não necessáriamente, afeta outras partes do mesmo sistema. Além disso, os componentes do projeto podem ser desenvolvidos paralelamente;

2. Fácil de manter: Mais fluidez no processo de atualização de códigos;

3. Extensível: É possível "acoplar" novas funcionalidades ao sistema a partir de objetos já criados. Se seguirmos o contexto da criação de um carro poderíamos entender da seguinte forma - criamos uma classe carro para representar o carro. 

Dessa forma é possível utilizar essa mesma classe para criar um sistema que possa trabalhar com vendas de carros. Novamente, podemos utilizar para criar um sistema que funcione para agendar tarefas para reparo de carros, oficinas mecânicas e entre outros projetos.


# Classe e Objeto

O que é um objeto ? É algo material ou abstrato que pode ser percebido pelos sentidos, e descrito por meio de duas características, comportamentos e estado atual (status), por exemplo o carro. 

Este cenário, entendemos que as características enumeradas são atributos de carro, como descrito abaixo:

- Características: marca, fabricante/montadora, cor, etc. Logo entendemos que as características enumeradas são atributos do carro;
- Comportamentos: Vamos considerar as funcionalidades que citamos acima, como acelerar, frear, trocar de marcha e entre outros;
- Estado atual: É representado pelos valores dos atributos naquele momento que é analisado, então entendemos que é possível ter um modelo de carro, no qual pode possuir diferentes características e funcionalidades. Esse modelo - molde, formato - chamamos de classe, CARRO = OBJETO MOLDE = CLASSE

```python
class Funcionario: 
    'Classe base comum para todos os funcionários'
# A variável funcioCount é uma variável de classe cujo valor é compartilhado entre todas as instâncias dessa classe
# Isso pode ser acessado como funcionario.funcioCount de detro da classe ou fora dela

    funcioCont = 0
        
# O primeiro método __init__() é um método especial ou consultor
# chamado consultor de classe ou método de inicialização, que o python chama quando você cria uma nova intância dessa classe.


    
    def __init__(self, nome, salario):
        self.nome = nome 
        self.salario = salario
        Funcionario.funcioCont += 1 
        
    def exibirfuncioCount(self):
        print ('Total de Funcionários: %d' % Funcionario.funcioCont)
        
    def exibirFuncionario(self):
        print("Nome: ", self.nome, '| Salário: ', self.salario)
        
# Criando o objeto da classe funcionario

func1 = Funcionario('CAIO', 6000)
func2 = Funcionario('C3PO4', 2000)
    
func1.exibirFuncionario()
func2.exibirFuncionario()
print('Total de Funcionários:  %d' % Funcionario.funcioCont)
```

```python
# ATRIBUTOS E FUNÇÕES

hasattr(func1, 'salario') # Retorna verdadeiro se o atributo salario existir 
getattr(func1, 'salario') # Retorna o valor do atributo 'salario'
setattr(func1, 'salario', 7000) # Define o atributo salario em 7K
delattr(func1, 'salario') # Exclui o atributo salario 
```

# Atributos incorporados 

Toda classe de Python continua seguindo os atributos internos e eles podem ser acessados usando o operador de ponto como qualquer outro atributo.

-__dict__ = Dicionário que contem o espaço para nome da classe;
- __doc__ = Sequência de documentações da classse ou nenhuma, se não definida;
- __module__ = Nome do módulo no qual a classe está definida. Este atributo é "__main__" no modo interativo;
- __bases__ = Uma tupla possivelmente vaziz que contém as classes base, na ordem em que ocorreram na lista de classe base.
